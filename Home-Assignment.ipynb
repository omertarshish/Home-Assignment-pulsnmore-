# Data Exploration

It seems there might be a sinusoidal correlation, which suggests the experiment includes a wave. First, we look for anomalies and delete them

```python
# Define the list of Y variable column names
Y = ['Y1', 'Y2', 'Y3', 'Y4', 'Y5']

# Define a threshold to identify anomalies (you can adjust this value based on your data)
threshold = 2.5

# Convert the columns to numeric if they contain non-numeric values in df_cleaned
df_cleaned[Y] = df_cleaned[Y].apply(pd.to_numeric, errors='coerce')

# Check if all Y variables are present in the DataFrame
missing_columns = [col for col in Y if col not in df_cleaned.columns]
if missing_columns:
    raise ValueError(f"The following columns are not present in the DataFrame: {', '.join(missing_columns)}")

# Loop through all Y variables
for column_name in Y:
    # Calculate the Z-score for the current Y variable in df_cleaned
    z_scores = (df_cleaned[column_name] - df_cleaned[column_name].mean()) / df_cleaned[column_name].std()

    # Filter the DataFrame to keep only the rows where the absolute Z-score is above the threshold
    anomalies_df = df_cleaned[abs(z_scores) > threshold]

    # Print the rows containing anomalies for the current Y variable
    print(f"Anomaly rows for {column_name}:")
    print(anomalies_df)

    # Remove the anomaly rows from df_cleaned
    df_cleaned = df_cleaned[abs(z_scores) <= threshold]

# Reset the index to ensure it is consecutive after deletion
df_cleaned = df_cleaned.reset_index(drop=True)

# Print the DataFrame after removing anomalies
print("DataFrame after removing anomalies:")
print(df_cleaned)

# Print the DataFrame info to see data types and non-null counts
print(df_cleaned.info())
```

Cross-correlation and the inverse cross-correlation to identify any exponential correlations.

```python
# Perform feature engineering (if needed) on Y2, Y3, Y4, and Y5
Y2_sqrt = np.sqrt(df_cleaned['Y2'])
Y3_sqrt = np.sqrt(df_cleaned['Y3'])
Y4_sqrt = np.sqrt(df_cleaned['Y4'])
Y5_sqrt = np.sqrt(df_cleaned['Y5'])

# Calculate cross-correlations
# Assuming data is collected over time and Y1 is the target variable
cross_corr_Y1_Y2 = np.correlate(df_cleaned['Y1'] - np.mean(df_cleaned['Y1']), Y2_sqrt - np.mean(Y2_sqrt), mode='full') / (np.std(df_cleaned['Y1']) * np.std(Y2_sqrt) * len(df_cleaned['Y1']))
cross_corr_Y1_Y3 = np.correlate(df_cleaned['Y1'] - np.mean(df_cleaned['Y1']), Y3_sqrt - np.mean(Y3_sqrt), mode='full') / (np.std(df_cleaned['Y1']) * np.std(Y3_sqrt) * len(df_cleaned['Y1']))
cross_corr_Y1_Y4 = np.correlate(df_cleaned['Y1'] - np.mean(df_cleaned['Y1']), Y4_sqrt - np.mean(Y4_sqrt), mode='full') / (np.std(df_cleaned['Y1']) * np.std(Y4_sqrt) * len(df_cleaned['Y1']))
cross_corr_Y1_Y5 = np.correlate(df_cleaned['Y1'] - np.mean(df_cleaned['Y1']), Y5_sqrt - np.mean(Y5_sqrt), mode='full') / (np.std(df_cleaned['Y1']) * np.std(Y5_sqrt) * len(df_cleaned['Y1']))



# Plot cross-correlations
plt.figure(figsize=(16, 6))
plt.subplot(141)
plt.plot(cross_corr_Y1_Y2)
plt.title("Cross-correlation: Y1 vs. Y2")
plt.xlabel("Lag")

plt.subplot(142)
plt.plot(cross_corr_Y1_Y3)
plt.title("Cross-correlation: Y1 vs. Y3")
plt.xlabel("Lag")

plt.subplot(143)
plt.plot(cross_corr_Y1_Y4)
plt.title("Cross-correlation: Y1 vs. Y4")
plt.xlabel("Lag")

plt.subplot(144)
plt.plot(cross_corr_Y1_Y5)
plt.title("Cross-correlation: Y1 vs. Y5")
plt.xlabel("Lag")

plt.tight_layout()
plt.show()



```

Linear regression of the correlation between the combined values of Y2-Y5 and Y1

```python
# Calculate the differences between Y1 and Y2, Y3, Y4, Y5
df_cleaned['Y1_Y2_diff'] = df_cleaned['Y1'] - df_cleaned['Y2']
df_cleaned['Y1_Y3_diff'] = df_cleaned['Y1'] - df_cleaned['Y3']
df_cleaned['Y1_Y4_diff'] = df_cleaned['Y1'] - df_cleaned['Y4']
df_cleaned['Y1_Y5_diff'] = df_cleaned['Y1'] - df_cleaned['Y5']

# Calculate the mean of Y2, Y3, Y4, Y5 in each row
df_cleaned['mean_Y2_Y5'] = df_cleaned[['Y2', 'Y3', 'Y4', 'Y5']].mean(axis=1)

# Calculate the difference between Y1 and the mean of Y2, Y3, Y4, Y5
df_cleaned['Y1_mean_Y2_Y5_diff'] = df_cleaned['Y1'] - df_cleaned['mean_Y2_Y5']

# Create separate scatter plots for each relationship and perform linear regression
plt.figure(figsize=(15, 10))

def plot_and_regression(x, y, subplot_index, title):
    plt.subplot(subplot_index)
    plt.scatter(x, y)
    plt.title(title)
    plt.xlabel("Y1")
    plt.ylabel(title)

    # Perform linear regression
    X = np.vstack([x, np.ones(len(x))]).T
    slope, intercept = np.linalg.lstsq(X, y, rcond=None)[0]

    # Plot the regression line
    plt.plot(x, slope * x + intercept, color='red', label=f"Y1 = {slope:.2f} * {title} + {intercept:.2f}")
    plt.legend()

    # Print stats
    print(f"Linear Regression: {title}")
    print(f"Equation: Y1 = {slope:.2f} * {title} + {intercept:.2f}")
    print(f"R-squared: {r_squared(x, y, slope, intercept):.2f}")
    print()

    # Return the slope and intercept for later use
    return slope, intercept

def r_squared(x, y, slope, intercept):
    y_pred = slope * x + intercept
    residual = y - y_pred
    ss_residual = np.sum(residual ** 2)
    ss_total = np.sum((y - np.mean(y)) ** 2)
    r_squared = 1 - (ss_residual / ss_total)
    return r_squared

# Plot and perform linear regression for Y1 - Y2
slope_Y1_Y2, intercept_Y1_Y2 = plot_and_regression(df_cleaned['Y1_Y2_diff'], df_cleaned['Y1'], 231, "Y2")

# Plot and perform linear regression for Y1 - Y3
slope_Y1_Y3, intercept_Y1_Y3 = plot_and_regression(df_cleaned['Y1_Y3_diff'], df_cleaned['Y1'], 232, "Y3")

# Plot and perform linear regression for Y1 - Y4
slope_Y1_Y4, intercept_Y1_Y4 = plot_and_regression(df_cleaned['Y1_Y4_diff'], df_cleaned['Y1'], 233, "Y4")

# Plot and perform linear regression for Y1 - Y5
slope_Y1_Y5, intercept_Y1_Y5 = plot_and_regression(df_cleaned['Y1_Y5_diff'], df_cleaned['Y1'], 234, "Y5")

# Plot and perform linear regression for Y1 - Mean(Y2, Y3, Y4, Y5)
slope_Y1_mean_Y2_Y5, intercept_Y1_mean_Y2_Y5 = plot_and_regression(df_cleaned['Y1_mean_Y2_Y5_diff'], df_cleaned['Y1'], 235, "Mean(Y2, Y3, Y4, Y5)")

# Compute the combined slope and intercept
combined_slope = (slope_Y1_Y2 + slope_Y1_Y3 + slope_Y1_Y4 + slope_Y1_Y5 + slope_Y1_mean_Y2_Y5) / 5
combined_intercept = (intercept_Y1_Y2 + intercept_Y1_Y3 + intercept_Y1_Y4 + intercept_Y1_Y5 + intercept_Y1_mean_Y2_Y5) / 5

# Print the combined slope and intercept
print("Combined Slope:", combined_slope)
print("Combined Intercept:", combined_intercept)
```

Ranking the new devices and conducting a test to check whether the new devices fail the tests,

```python
df_new_devices['mean_Y2_Y5'] = df_new_devices[['Y2', 'Y3', 'Y4', 'Y5']].mean(axis=1)

# Calculate the difference between Y1 and the mean of Y2, Y3, Y4, Y5 for the new devices
df_new_devices['Y1_mean_Y2_Y5_diff'] = df_new_devices['Y1'] - df_cleaned['mean_Y2_Y5']

# Upredict the Y1 values for the new devices
X_new_devices = df_new_devices['Y1_mean_Y2_Y5_diff'].values.reshape(-1, 1)  # Reshape to a 2D array
Y_predicted_new_devices = (combined_slope * X_new_devices) + combined_intercept

# Add the 'predicted_Y1' column to the new_devices_data DataFrame
df_new_devices['predicted_Y1'] = Y_predicted_new_devices

# Rank the devices based on the absolute differences between predicted_Y1 and actual Y1 values.
df_new_devices['absolute_difference'] = abs(df_new_devices['predicted_Y1'] - df_new_devices['Y1'])

# Sort the DataFrame in ascending order of absolute_difference values
df_new_devices_ranked = df_new_devices.sort_values(by='absolute_difference', ascending=True)

#Determine which devices fail the test based on a specified threshold for the absolute differences.
failure_threshold = 0.1  # Adjust  threshold
df_new_devices_ranked['test_result'] = df_new_devices_ranked['absolute_difference'] > failure_threshold

# Print the ranked DataFrame for new devices
print(df_new_devices_ranked)
```

The test assesses the accuracy of a linear regression model in predicting Y1 values for new devices. It calculates the absolute difference between predicted and actual Y1 values and ranks the devices based on these differences. Devices with an absolute difference greater than 0.1 are considered to fail the test.
